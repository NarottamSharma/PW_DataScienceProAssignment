# 2. Given a recurrence relation, solve it using the substitution method:

```python
a. T(n) = T(n-1) + c

The recurrence relation is given as:

T(n) = T(n-1) + c

We can start by expanding this relation:

T(n) = T(n-1) + c = T(n-2) + c + c = T(n-3) + c + c + c 

Continuing this process, we get:

T(n) = T(n-k) + kc

We know that the base case is T(1), so n-1=k  ,k=n-1, and substituting this in we get:

T(n) = T(1) + c(n-1)

So, the solution to the given recurrence relation is T(n) = T(1) + c(n-1).
```

```python
b. T(n) = 2T(n/2) + n

The recurrence relation is given as:

T(n) = 2T(n/2) + n

We can start by expanding this relation:

T(n) = 2T(n/2) + n = 2[2T(n/4) + n/2] + n = 2^2T(n/2^2) + 2n

Continuing this process, we get:

T(n) = 2^kT(n/2^k) + kn

We know that the base case is T(1), so n/2^k=1 => k=log_2{n}, and substituting this in we get:

T(n) = nT(1) + nlog_2{n}

So, the solution to the given recurrence relation is T(n) = nT(1) + nlog_2{n}. 
```

# 3. Given a recurrence relation, solve it using the recursive tree approach:

```python
a. T(n) = 2T(n-1) +1

To solve the given recurrence relation T(n) = 2T(n-1) + 1 using the recursive tree approach, we can visualize the recursion as a tree. Each node in the tree represents a subproblem, and the edges represent the recursive calls. Here's how the recursive tree looks for this recurrence relation:


      T(n)
      /  \
  T(n-1)  T(n-1)
   / \      / \
T(n-2) T(n-2) ...
  / \      / \
...  ...  ... ...


At each level of the tree, the problem size decreases by 1, and each node has two children corresponding to the recursive calls with (n-1) as the argument.

Now, let's write down the recursive relation for the tree:


T(n) = 2T(n-1) + 1


T(n-1) = 2T(n-2) + 1

T(n-2) = 2T(n-3) + 1


T(2) = 2T(1) + 1

Now, let's substitute these relations back into the original one:

T(n) = 2[2T(n-2) + 1] + 1



T(n) = 2^2T(n-2) + 2 + 1


Similarly,


T(n) = 2^3T(n-3) + 2^2 + 2 + 1


Continuing this pattern, we get:


T(n) = 2^kT(n-k) + 2^{k-1} + 2^{k-2} + .... + 2^2 + 2 + 1


Now, we know that the base case is (T(1)), so when (n-k = 1), we have (k = n-1). Substituting this back:


T(n) = 2^{n-1}T(1) + 2^{n-2} + 2^{n-3} + .... + 2^2 + 2 + 1


Finally, we can simplify this using the geometric series formula:

T(n) = 2^{n-1}T(1) + 2^{n-1} - 1

So, the solution to the given recurrence relation using the recursive tree approach is T(n) = 2^{n-1}T(1) + 2^{n-1} - 1.
```
